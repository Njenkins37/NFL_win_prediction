```python
import pandas as pd
import numpy as np
```

```python
team_data = pd.read_csv("week2.csv")
team_data.dropna(inplace=True)


team_data
```

```python
def create_team_data (data, name, number):
    condition = data['Team Number'] == number
    filtered_teams = data.loc[condition]
    name = filtered_teams[['Rushing Yards', 'Passing Yards', 'Time Of Possession',
                        'Rushing Yards allowed', 'Passing Yards Allowed','Penalty Yards', 'Turnovers']].mean()
    name_array = name.values
    return name_array
```

```python
team_list = [
    'arizona', 
    'atlanta',
    'baltimore',
    'buffalo', 
    'carolina', 
    'chicago', 
    'cincinatti', 
    'cleveland', 
    'dallas', 
    'denver', 
    'detroit', 
    'green bay', 
    'houston',
    'indianapolis', 
    'kansas city', 
    'chargers',
    'rams'
    'jacksonville', 
    'miami', 
    'minnesota', 
    'new england', 
    'new orleans', 
    'giants', 
    'jets', 
    'raiders',
    'philadelphia', 
    'pittsburgh', 
    'san francisco', 
    'seattle', 
    'tampa bay', 
    'tennessee', 
    'washington'
]
```

```python
team_dict = {}
number = 0
for team in team_list: 
    name = team_list[number]
    team_value = create_team_data(team_data, name, number + 1)
    team_dict[name]= team_value
    number = number + 1

print(team_dict)
```

```python
team_data = team_data.drop("Team Number", axis = 1)
```

```python
team_data= team_data.drop("Year", axis = 1)

team_data
```

```python
X = team_data.drop("Win", axis = 1)
X
```

```python
y = team_data["Win"]
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2
                                                   )
```

```python
from sklearn.ensemble import RandomForestClassifier
np.random.seed(42)

model = RandomForestClassifier()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```
